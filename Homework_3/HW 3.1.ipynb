```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("hw2_1.csv", sep=';')
data
```

```python
plt.bar(data['2019'][:27], data['Регіон'][:27],)
plt.show()
```

```python
plt.title('Коефіцієнт народжуваності в України (2019)', fontsize=12)
plt.plot(data['2019'][:27], data['Регіон'][:27],)
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(15,5))
axs.bar(data['Регіон'][:27], data['2019'][:27], color=["b", "r", "y", "black", "g", "m"])
plt.ylabel('Народжуваність на рік')
plt.title('Народжуваність в Україні за регіонами (2019)')
for tick in axs.get_xticklabels():
            tick.set_rotation(55)
plt.grid(axis='y')
plt.show()
```
