```python
%matplotlib inline
import matplotlib.pyplot as plt
import random

dot_1 = [0, 0]
dot_2 = [5000, 10000]
dot_3 = [10000, 0]

x = [0, 5000, 10000]  # x -   
y = [0, 10000, 0]  # y -   


def audit(x_dot, y_dot):
    a = 5000 * y_dot - 10000 * x_dot
    b = 10000 * x_dot + 5000 * y_dot - 100000000
    c = -10000 * y_dot

    if (a >= 0 and b >= 0 and c >= 0) or (a <= 0 and b <= 0 and c <= 0):
        return True
    else:
        return False


def first_dot():
    while True:
        x_dot = random.randint(0, 10000)
        y_dot = random.randint(0, 10000)

        if audit(x_dot, y_dot):
            x.append(x_dot)
            y.append(y_dot)
            return x_dot, y_dot


def half_dot(x_dot, y_dot):

    r_dot = random.choice([dot_1, dot_2, dot_3])

    x_dot = (x_dot + r_dot[0])/2
    y_dot = (y_dot + r_dot[1])/2

    if audit(x_dot, y_dot):
        return x_dot, y_dot


def main():
    x_dot, y_dot = first_dot()
    i = 0

    while i < 50000:
        i += 1
        x_dot, y_dot = half_dot(x_dot, y_dot)

        x.append(x_dot)
        y.append(y_dot)

    visualization()


def visualization():
    fig, ax = plt.subplots()
    

    ax.scatter(x, y, c = 'yellow', s = 1)  #    

    ax.set_facecolor('blue')  #   Axes
    ax.set_title(' ')  #   Axes

    fig.set_figwidth(10)  #  
    fig.set_figheight(10)  #  "Figure"

    plt.show()


if __name__ == "__main__":
    main()
```
