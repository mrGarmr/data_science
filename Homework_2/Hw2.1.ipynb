* Вывести первые строки таблицы с помощью метода head

```python
import numpy as np

import pandas as pd

tmp = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C", match = 'Коефіцієнт народжуваності в регіонах України', thousands='.', decimal=',')
table = tmp[0]
table.head()
```

* Определите количество строк и столбцов в датафрейме (атрибут shape)

```python
print(f'Количество строк: {table.shape[0]} и столбцов: {tmp[0].shape[1]}')
```

* Замените в таблице значения "—" на значения NaN

```python
table = table.replace(["—"], "NaN")
table
```

* Определите типы всех столбцов с помощью dataframe.dtypes

```python
table.dtypes
```

* Замените типы не числовых колонок на числовые. Подсказка - это колонки где находился символ "—"

```python
table['2014'] = table['2014'].astype(float)
table['2019'] = table['2019'].astype(float)
table.dtypes
```

* Посчитайте, какая доля пропусков содержится в каждой колонке (используйте методы isnull и sum)

```python
table.isnull().sum()
```

* Удалите из таблицы данные по всей стране, последняя строчка таблицы

```python
table.drop([27], axis=0)
```

* Замените отсутствующие данные в столбцах средними значениями по этим столбцам (метод fillna)

```python
table = table.fillna(table.mean(numeric_only=True), axis=0)
table
```

* Получите список регионов, где уровень рождаемости в 2019 году был выше среднего по Украине

```python
high_level = table.loc[table['2019'] > table['2019'][27]]
high_level
```

* В каком регионе была самая высокая рождаемость в 2014 году?


```python
result=table.nlargest(1,['2014'])
result
```

* Постройте столбцовую диаграмму рождаемости по регионам в 2019 году

```python
import matplotlib.pyplot as plt


table[:27].plot.bar(x='Регіон', y=['1950','1960','1970','1990','2000','2012','2014','2019'])
plt.show()
```

```python

table[:27].plot.barh(x='Регіон', y=['1950','1960','1970','1990','2000','2012','2014','2019'])
plt.title('Diagram of birth')
plt.xlabel('Коефіцієнт')
plt.ylabel('Регіон')
plt.show()
```
